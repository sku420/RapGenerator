```python
#read
def loadtxt(filename):
    with open(filename,encoding="utf-8") as f:
        return f.read().lower()
```

```python
data=loadtxt("data.txt")
```

```python
def transitionTable(data, k=4):
    T={}
    #work
    for i in range(0,len(data)-k):
        X=data[i:i+k]
        Y=data[i+k]
        if T.get(X) is None:
            T[X]={}
            T[X][Y]=1
        elif T[X].get(Y) is None:
            T[X][Y]=1
        else:
            T[X][Y]+=1
    return T
```

```python
test=transitionTable(data)
```

```python
def probability(T):
    for kx in T.keys():
        norm_factor=sum(T[kx].values()) #denominator
        for val in T[kx].keys():
            T[kx][val]/=norm_factor
    return T
```

```python
test_probab=probability(test)
```

```python
test_probab
```

```python
def predict(string, T):
    if T.get(string) is None:
        return " "
    next_char = list(T[string].keys())
    probab = list(T[string].values())
    return np.random.choice(next_char, p=probab)
```

```python
sentence = "कौन "
for i in range(1000):
    next_char = predict(sentence[-4:], test_probab)
    sentence+=next_char
print(sentence)
```
